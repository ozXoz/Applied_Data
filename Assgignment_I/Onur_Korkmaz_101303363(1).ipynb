```python
movies = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head() 
```

```python
movies.shape 
```

```python
(movies['duration']>=200)
```

```python
movies['genre']=='Drama'
```

```python
(movies['duration']>=200) & (movies['genre']=='Drama')
```

```python
movies[(movies['duration']>=200) & (movies['genre']=='Drama')] 
```

```python
movies[(movies['duration']>=200) | (movies['genre']=='Drama')] 
```

```python
movies[(movies['duration']>=200) | (movies['genre']=='Drama')].shape
```
